```python
import numpy as np
import matplotlib.pyplot as plt
import random
from numpy.random import random as rnd
```

```python
#Punto 1


def fibonacci(n):
    """Genera los primeros n números de la serie de Fibonacci."""
    fib_series = [1, 1]  # Comienza con 1 y 1
    for _ in range(n - 2):
        fib_series.append(fib_series[-1] + fib_series[-2])
    return fib_series

def plot_fibonacci(n):
    """Dibuja un gráfico de líneas con los primeros n números de Fibonacci."""
    fib_numbers = fibonacci(n)

    plt.figure(figsize=(8, 5))  # Tamaño del gráfico
    plt.plot(range(1, n + 1), fib_numbers, marker='o', linestyle='-', color='b', label="Fibonacci")

    plt.xticks(range(1, n + 1))  # Ajuste de valores en el eje X
    plt.title(f'Serie de Fibonacci (Primeros {n} términos)')  # Título del gráfico
    plt.xlabel('Índice en la serie')
    plt.ylabel('Valor de Fibonacci')
    plt.legend()  # Muestra la leyenda
    plt.grid(axis='both', linestyle='--', alpha=0.6)  # Cuadrícula suave
    plt.show()

# Solicitar el número de términos
n = int(input("Ingrese el número de términos de Fibonacci a generar: "))
if n > 0:
    plot_fibonacci(n)
else:
    print("Por favor, ingrese un número mayor a 0.")
```

```python
#Punto 2
def analyze_temperatures(temperatures):
    mean_temp = np.mean(temperatures)
    median_temp = np.median(temperatures)
    std_dev_temp = np.std(temperatures)

    print(f"Media: {mean_temp:.2f}°C")
    print(f"Mediana: {median_temp:.2f}°C")
    print(f"Desviación estándar: {std_dev_temp:.2f}°C")

    months = np.arange(1, 13)
    monthly_means = [np.mean(temperatures[i::12]) for i in range(12)]  # Promedio mensual

    plt.figure(figsize=(8, 5))
    plt.plot(months, monthly_means, marker='o', linestyle='-', color='b', label="Temp. Media")
    plt.title('Temperatura Media Mensual')
    plt.xlabel('Meses')
    plt.ylabel('Temperatura (°C)')
    plt.xticks(months)  # Asegurar que el eje x muestre 1-12
    plt.grid(True, linestyle='--', alpha=0.6) #el doble guin define que la linea de la cuadricula es entrecortada, el alpha=0.6 define la transparencia
    plt.legend()  #muestra la temp.media en el lado derecho superior
    plt.show()
temperatures = [25.6, 26.1, 24.8, 27.3, 28.4, 26.7, 25.9, 29.1, 30.2, 27.8, 24.3, 23.9,
    26.5, 28.0, 27.4, 29.3, 30.5, 26.8, 25.7, 24.6]

analyze_temperatures(temperatures)
```

```python
#Punto 3

def simulate_dice_rolls():
    rolls = [random.randint(1, 6) for _ in range(1000)]  # Genera 1000 valores entre 1 y 6 (simulación de dado)
    values, counts = np.unique(rolls, return_counts=True)  # Obtiene valores únicos y su frecuencia
    plt.bar(values, counts, color='g', alpha=0.7)
    plt.xticks(values)
    plt.title('Frecuencia de lanzamientos de un dado (1000 tiradas)')
    plt.xlabel('Número del dado')
    plt.ylabel('Frecuencia')
    plt.grid(axis='y', linestyle='--', alpha=0.6)  #axis determina que las lineas solo aparecen en el eje y
    plt.show()

simulate_dice_rolls()
```

```python
#Punto 7

VelocidadInicial = 56 # @param {"type":"slider","min":0,"max":100,"step":1}
Angulo = 19 # @param {"type":"slider","min":0,"max":90,"step":1}

AnguloRad = np.radians(Angulo)

vx = VelocidadInicial * np.cos(AnguloRad)
vy = VelocidadInicial * np.sin(AnguloRad)

tiempoVuelo = (2 * vy) / 9.8

t = np.linspace(0, tiempoVuelo,100)
x = vx * t
y = vy * t - 0.5 * 9.8 * t**2

plt.plot(x, y)
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlim(-1,50)
plt.ylim(-1,50)
plt.show()
```

```python
#Punto 9

n = 100000

inicio_x = 0.5
inicio_y = 0.65

puntos_de_x = []
puntos_de_y = []

numeros_de_x = 0
numeros_de_y = 0

puntos_de_x.append(inicio_x)
puntos_de_y.append(inicio_y)

lista = []
lista = rnd(n)

veticex1 = 0
verticey1 = 0
verticex2 = 0.5
verticey2 = np.sqrt(3)/2
verticex3 = 1
verticey3 = 0

for i in range(n):
  if (lista[i] <1/3):
    verticex = veticex1
    verticey = verticey1
  elif (1/3<lista[i]<2/3):
    verticex = verticex2
    verticey = verticey2
  elif (lista[i]>2/3):
    verticex = verticex3
    verticey = verticey3

  if (i==0):
    numeros_de_x = (verticex+numeros_de_x)/2
    numeros_de_y = (verticey+numeros_de_y)/2
  else:
    numeros_de_x = (verticex+numeros_de_x)/2
    numeros_de_y = (verticey+numeros_de_y)/2
  puntos_de_x.append(numeros_de_x)
  puntos_de_y.append(numeros_de_y)

plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(y=0, color='b')
plt.axvline(x=0, color='b')
plt.plot(puntos_de_x, puntos_de_y,".",ms = 0.25)
plt.show()
```
